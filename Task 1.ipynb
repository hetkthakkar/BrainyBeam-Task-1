### Description: Create animated plots using Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
```

```python
data=pd.read_csv("D://Study//Parul University//PUStudy//#Placement//BrainyBeam Info-Tech Pvt Ltd-Ahemdabad//Task 1//code//tips.csv")
```

### Line Chart

```python
# Sort by total_bill
data_sorted = data.sort_values(by='total_bill')  

# Prepare the data
x = data_sorted['total_bill'].values
y = data_sorted['tip'].values

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_facecolor('mistyrose')  # background color
ax.set_title('Line Chart: Tip Amount vs Total Bill', fontsize=14, fontweight='bold')
ax.set_xlabel('Total Bill ($)', fontsize=12)
ax.set_ylabel('Tip Amount ($)', fontsize=12)

# Initialize the line
line, = ax.plot([], [], lw=2, color='deeppink', label='Tips')

# Set axis limits
ax.set_xlim(x.min(), x.max())
ax.set_ylim(0, y.max())

# Gridlines
ax.grid(color='gray', linestyle=':', linewidth=0.5)

# Legend
ax.legend(loc='upper left', fontsize=10)

# Update function for animation
def update(frame):
    line.set_data(x[:frame], y[:frame])  # Update line data
    return line,

# Create the animation
ani = FuncAnimation(fig, update, frames=len(x), interval=50, blit=True)

# Save animation as GIF
ani.save("line_chart.gif", writer=PillowWriter(fps=30))
```
